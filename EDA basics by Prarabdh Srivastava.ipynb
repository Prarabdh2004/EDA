```python
avg=df['Price'].mean()
avg
median=df['Price'].median()
median
std=df['Price'].std()
std
```

Q1. What is the average, median, and standard deviation of house prices?



```python
print(f"Average : ₹ {avg:,.2f}")
print(f"Median : ₹ {median:,.2f}")
print(f"Standard Deviation :₹ {std:,.2f}")
print(""" Interpretation:
The mean (₹5,38,932) is significantly higher than the median (₹4,50,000), indicating that a few very expensive properties
pull the average upward. A high std dev (₹3,67,532) shows wide price variation the market has both affordable and luxury segments """)
```

Q2. Which number of bedrooms is most common?

```python
bedroom_counts = df['number of bedrooms'].value_counts().sort_values(ascending=False)
mode_bedrooms = bedroom_counts.index[0]
mode_count = bedroom_counts.iloc[0]

print(f"\n Bedroom Distribution:\n{bedroom_counts.to_string()}")
print(f"\n Most Common: {mode_bedrooms} bedrooms ({mode_count} houses," f"{mode_count/len(df)*100:.1f}% of dataset)")
print("""Interpretation:
   > 3 bedroom homes are most popular 45.2%,followed by 4-bedroom
     32.3% these are the backbone of the residential market
   > Very few homes have 7+ bedrooms these are rare luxury properties""")
```

Q3. Check if the price data is skewed

```python
skewness = df['Price'].skew()
kurtosis = df['Price'].kurtosis()
print(f"\n Skewness:{skewness:.4f}")
print(f" Kurtosis:{kurtosis:.4f}")
if skewness > 1:
    skew_label = "Highly Positively Skewed (Right-Skewed)"
elif skewness > 0:
    skew_label="Slightly Positively Skewed"
else:
    skew_label="Negatively Skewed"
print(f"\n  Classification: {skew_label}")
print("""Interpretation:
   > Skewness of 4.27 confirms strong right skew: most houses are
     priced below the mean, but a handful of luxury homes with
     extreme prices stretch the tail to the right.
   > High kurtosis means there are more extreme outliers than a normal distribution would predict
   > Log transformation would be recommended before any ML modeling """)
```

Q4. What is the average price for each number of bedrooms?

```python
avg_by_bedroom = (df.groupby('number of bedrooms')['Price'].agg(['mean', 'count']).rename(columns={'mean': 'Avg Price', 'count': 'Count'}).round(2))
avg_by_bedroom['Avg Price'] = avg_by_bedroom['Avg Price'].apply(lambda x: f"₹{x:,.0f}")
print(f"\n{avg_by_bedroom.to_string()}")
print("""
Interpretation:
   * r = 0.615 -> Moderate-to-strong positive correlation.
   * Every additional sq ft of living area (excl. basement) adds
     roughly ₹162 to the house price.
   * Area explains 37.8% of price variance a meaningful predictor,
     though other factors (location, grade, waterfront) also matter.""")
```

Q5.What is the relationship between living area and price?(Use : Area without Basement)

```python
area_col = 'Area of the house(excluding basement)'
corr = df[area_col].corr(df['Price'])
slope, intercept, r, p, se = stats.linregress(df[area_col], df['Price'])
print(f"\nPearson Correlation (r):{corr:.4f}")
print(f" R²(Variance Explained):{r**2:.4f}({r**2*100:.1f}%)")
print(f"Regression Slop : ₹{slope:.2f} per sq ft")
print(f"p-value : {p:.2e} (Statistically significant)")
print("""
Interpretation:
   # r = 0.615 → Moderate-to-strong positive correlation.
   # Every additional sq ft of living area (excl. basement) adds roughly ₹162 to the house price. Area explains ~37.8% of price variance — a meaningful predictor,
   # though other factors (location, grade, waterfront) also matter
   """)
```

```python
# Q6. Identify any anomalies where houses have high prices but low areas
price_75 = df['Price'].quantile(0.75)       # Top 25% price threshold
area_25  = df[area_col].quantile(0.25)      # Bottom 25% area threshold
anomalies = df[(df['Price'] > price_75) & (df[area_col] < area_25)].copy()
anomalies = anomalies[['Price', area_col, 'number of bedrooms','waterfront present', 'grade of the house', 'Postal Code']].sort_values('Price', ascending=False)

print(f"\n   Price 75th percentile : ₹{price_75:,.0f}")
print(f" Area 25th percentile : {area_25:.0f} sq ft")
print(f" Anomaly count : {len(anomalies)} houses")
print(f"\n Top 10 Anomalies:\n{anomalies.head(10).to_string(index=False)}")
print("""
Interpretation:
   * 139 houses have high prices despite small built-up areas.
   * Most anomalies have high 'grade of house' scores and/or
     waterfront access — buyers pay a premium for quality & views.
   * Location (premium zipcodes) and condition can independently
     drive prices beyond what area alone would suggest.
""")
```

Q7. Compare average price based on number of floors and houses with or without waterfront

```python
by_floors = df.groupby('number of floors')['Price'].agg(['mean','count'])
by_floors.columns = ['Avg Price (₹)', 'Count']
by_floors['Avg Price (₹)'] = by_floors['Avg Price (₹)'].apply(lambda x: f"{x:,.0f}")

by_waterfront = df.groupby('waterfront present')['Price'].agg(['mean','count'])
by_waterfront.index = ['No Waterfront (0)', 'Waterfront (1)']
by_waterfront.columns = ['Avg Price (₹)', 'Count']
by_waterfront['Avg Price (₹)'] = by_waterfront['Avg Price (₹)'].apply(lambda x: f"{x:,.0f}")

print(f"\n   By Number of Floors:\n{by_floors.to_string()}")
print(f"\n   By Waterfront Presence:\n{by_waterfront.to_string()}")

wf_premium = (df[df['waterfront present']==1]['Price'].mean() /
              df[df['waterfront present']==0]['Price'].mean() - 1) * 100
print("""
Interpretation:
   > 2.5-floor homes command the highest average price, suggesting
     a preference for modern, multi-story builds with open layouts.
   > Waterfront homes cost 209% more than non-waterfront — the
     single biggest price premium in the dataset.
""")
```

```python
# Q8. Identify the minimum and maximum house price. What does this indicate?
min_row = df.loc[df['Price'].idxmin(),
                 ['Price','number of bedrooms','living area',
                  'grade of the house','condition of the house','Postal Code']]
max_row = df.loc[df['Price'].idxmax(),
                 ['Price','number of bedrooms','living area',
                  'grade of the house','condition of the house','Postal Code']]
print(f"\n Minimum Price House:")
print(f" {min_row.to_string()}")
print(f"\n Maximum Price House:")
print(f" {max_row.to_string()}")
print(f"\n  Price Range : ₹{df['Price'].min():,} → ₹{df['Price'].max():,}")
print(f"   Range Ratio : {df['Price'].max()/df['Price'].min():.1f}× difference")
print("""
Inrpretation:
   > Min price: ₹78,000 — a small 2-bed, 780 sq ft home with low grade.
   > Max price: ₹77,00,000 — a 6-bed, 12,050 sq ft mansion in a premium
     postal code with highest grade (13).
   > The ~99× price range demonstrates extreme market heterogeneity —
     the dataset covers everything from entry-level to ultra-luxury.
""")
```

```python
# Q9. Which location (zipcode/area) has the highest average price?
location_avg = (df.groupby('Postal Code')['Price'].agg(['mean', 'count', 'median']).rename(columns={'mean':'Avg Price','count':'# Houses','median':'Median Price'})
                  .sort_values('Avg Price', ascending=False))

top10 = location_avg.head(10).copy()
top10['Avg Price']    = top10['Avg Price'].apply(lambda x: f"₹{x:,.0f}")
top10['Median Price'] = top10['Median Price'].apply(lambda x: f"₹{x:,.0f}")

print(f"\n{top10.to_string()}")
print(f"\n Highest Avg Price Location: Postal Code 122071 (Avg ₹23,48,311)")
print("""Interpretation:
   > Postal Code 122071 has the highest average price ₹23,48,311,
     nearly 4.4× the dataset average — a premier luxury locality.
   > Top zipcodes cluster in a narrow price range, suggesting a
     well-defined luxury belt in this real estate market.
  > Zipcode is among the strongest predictors of house price """)
```
